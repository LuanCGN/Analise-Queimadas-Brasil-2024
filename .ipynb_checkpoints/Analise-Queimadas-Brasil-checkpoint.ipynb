<h2>2.1. Grafico de barras -  Historico queimadas Brasil 1998 - 2024 </h2>

<h4>Nessa parte faremos uma vizualiçao do total de focos de queimadas dos anos de 98 até 2024 e fazer uma comparaçao entre o ano com maior foco de queimada e 2024</h4>

```python
# Normalizar os totais para gerar cores
norm = plt.Normalize(df_queimadashistoric['Total'].min(), df_queimadashistoric['Total'].max())
colors = plt.cm.Reds(norm(df_queimadashistoric['Total']))

# Suavizar as cores utilizando uma interpolação
colors = [(r * 0.8 + 0.2, g * 0.8, b * 0.8) for r, g, b, a in colors]

# Criar o gráfico de barras vertical
plt.figure(figsize=(10, 6))
plt.bar(df_queimadashistoric['Ano'], df_queimadashistoric['Total'], color=colors)

# Adicionar título e rótulos aos eixos
plt.title('Total de Queimadas por Ano', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Total', fontsize=12)

# Exibir o gráfico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<h2> 2.5. Gráfico Pizza - Focos por bioma - 2024:</h2>

<h4>Nesse DataFrame cada linha representa um registro de ocorrência de queimadas, por isso os gráficos onde mostra focos especificos estão sendo criados como countplots, que por padrão, conta quantas vezes cada categoria (neste caso, cada bioma) aparece no DataFrame.</h4>

```python
# Contar o número de queimadas por bioma
contagem_biomas = df_queimadas['Bioma'].value_counts()

# Definindo cores específicas para cada bioma
cores_biomas = {
    'Cerrado': 'darkorange',  # Laranja escuro
    'Amazônia': 'darkgreen',   # Verde escuro
    'Mata Atlântica': 'blue',   # Azul
    'Caatinga': 'yellow',       # Amarelo
    'Pantanal': 'lightseagreen', # Verde azulado
    'Pampa': 'red'              # Vermelho
}

# Aplicar as cores na ordem correta
paleta_cores = [cores_biomas[bioma] for bioma in contagem_biomas.index]

# Criando o gráfico de pizza
plt.figure(figsize=(18, 12))
plt.pie(
    contagem_biomas, 
    colors=paleta_cores, 
    startangle=90, 
    wedgeprops=dict(edgecolor='w')  # Para bordas brancas entre os setores
)

# Calcular as porcentagens
porcentagens = (contagem_biomas / contagem_biomas.sum()) * 100

# Criar rótulos para a legenda com porcentagens
rotulos_legenda = [f'{bioma}: {porc:.1f}%' for bioma, porc in zip(contagem_biomas.index, porcentagens)]

# Adicionando a legenda ao lado do gráfico, afastando-a
plt.legend(
    labels=rotulos_legenda,
    loc='upper left',  # Localização inicial da legenda
    bbox_to_anchor=(1, 1),  # Ajusta a posição da legenda
    title='Biomas',
    frameon=False,
    fontsize=20,
    title_fontsize='20',
    borderaxespad=1.5  # Aumenta o espaço entre a legenda e o gráfico
)

# Exibindo o gráfico
plt.axis('equal')  # Para garantir que a pizza seja desenhada como um círculo
plt.title('Distribuição de Queimadas por Bioma', fontsize=20)
plt.show() 
```

```python
biomasCasos = df_queimadas.groupby('Bioma')['Bioma'].count().sort_values(ascending=False)
biomasCasos.head(5)
```

<h2>2.6. Grafico linha - Relação de dias sem chuvas com os biomas - 2024:</h2>

<h4>Este gráfico de linha mostra a média de dias sem chuva em diferentes biomas. Ele agrupa os dados por bioma, calcula a média de dias sem precipitação para cada um, e depois exibe esses valores em um gráfico. O objetivo é visualizar a relação entre os biomas e o número médio de dias sem chuva, permitindo identificar quais biomas experimentam períodos mais longos ou curtos de seca.</h4>

```python
# Substitui vírgulas por pontos na coluna 'DiaSemChuva' e converte para tipo float
df_queimadas['DiaSemChuva'] = df_queimadas['DiaSemChuva'].replace(',', '.', regex=True).astype(float)
# Agrupa os dados por 'Bioma' e calcula a média da coluna 'DiaSemChuva' para cada bioma
df_mean_dias_sem_chuva = df_queimadas.groupby('Bioma')['DiaSemChuva'].mean().reset_index()
# A função groupby('Bioma') agrupa o DataFrame pelos biomas, e o mean() calcula a média dos valores de 'DiaSemChuva' dentro de cada bioma
# Cria uma figura com tamanho 8x6 polegadas
plt.figure(figsize=(8,6))
# Cria um gráfico de linha mostrando a relação entre Bioma e a média de dias sem chuva, com marcadores circulares ('o') e cor azul
sns.lineplot(x='Bioma', y='DiaSemChuva', data=df_mean_dias_sem_chuva, marker='o', color='b')
# Define o rótulo do eixo X como 'Bioma'
plt.xlabel('Bioma')
# Define o rótulo do eixo Y como 'Média de Dias sem chuva'
plt.ylabel('Média de Dias sem chuva')
# Define o título do gráfico
plt.title('Relação entre Dias Sem Chuva e Biomas (Média por Bioma)')
# Exibe o gráfico
plt.show()
```

```python
biomas_dias_sem_chuva = df_queimadas.groupby('Bioma')['DiaSemChuva'].sum().sort_values(ascending=False)
biomas_dias_sem_chuva.head(6)
```

<h2>2.7. Grafico linha/Barras - Relação entre Focos de Queimadas e Média de Dias Sem Chuva por Bioma:</h2>

```python
bioma_dias_focos = df_queimadas.groupby('Bioma').agg(
    media_dias_sem_chuva=('DiaSemChuva', 'mean'),  
    total_focos=('Bioma', 'count')  
).sort_values(by='total_focos', ascending=False)
fig, ax1 = plt.subplots(figsize=(8,6))
ax1.bar(bioma_dias_focos.index, bioma_dias_focos['total_focos'], color='b', label='Total de Focos de Queimadas')
ax1.set_xlabel('Bioma')
ax1.set_ylabel('Total de Focos de Queimadas', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax2 = ax1.twinx()
ax2.plot(bioma_dias_focos.index, bioma_dias_focos['media_dias_sem_chuva'], color='r', marker='o', label='Média de Dias Sem Chuva')
ax2.set_ylabel('Média de Dias Sem Chuva', color='r')
ax2.tick_params(axis='y', labelcolor='r')
plt.title('Relação entre Focos de Queimadas e Média de Dias Sem Chuva por Bioma')
plt.show()
```

<h2>2.8. Conclusão:</h2>
    <p>
        Em geral, a relação entre dias sem chuva e focos de queimadas varia bastante entre os biomas:
    </p>
    <ul>
        <li><strong>Cerrado</strong> apresenta uma forte correlação positiva, com um alto número de focos de queimadas associado a um grande número de dias sem chuva.</li>
        <li><strong>Amazônia</strong> tem um comportamento atípico devido ao valor negativo de dias sem chuva, o que pode indicar problemas de qualidade de dados ou que outras variáveis têm um impacto maior nas queimadas.</li>
        <li>Outros biomas, como <strong>Mata Atlântica</strong> e <strong>Caatinga</strong>, mostram uma correlação moderada, enquanto <strong>Pantanal</strong> e <strong>Pampa</strong> têm uma menor correlação, possivelmente devido a fatores ambientais específicos que limitam a ocorrência de queimadas.</li>
    </ul>
    <p>
        A relação entre dias sem chuva e focos de queimadas pode ser complexa e influenciada por uma combinação de fatores climáticos, ambientais e humanos.
    </p>

Técnicas agrícolas tradicionais também podem explicar a relação entre desmatamento e queimadas em estados como Rondônia e Acre. Em resumo, o desmatamento impulsiona as queimadas, com variações conforme as práticas econômicas regionais.</p>

<h1>4. Medidas para solucionar o problema</h1>

<h2>4.1. Estratégias de Combate às Queimadas:</h2>
<h3>4.1.1. Reforço na Fiscalização e Controle</h3>
    <ul>
        <li>Aumentar a fiscalização ambiental, utilizando tecnologias como satélites, drones e inteligência artificial para monitorar queimadas em tempo real e criar zonas de proteção integral.</li>
    </ul>

<h3>4.1.2. Promoção de Práticas Agrícolas Sustentáveis</h3>
    <ul>
        <li>Incentivar o uso de práticas agrícolas de baixo impacto, como agroecologia e plantio direto, além de programas de pagamento por serviços ambientais (PSA) para recompensar agricultores que preservam a vegetação.</li>
    </ul>

<h3>4.1.3. Recuperação de Áreas Degradadas</h3>
    <ul>
        <li>Investir no reflorestamento e na criação de corredores ecológicos para conectar áreas nativas e permitir a regeneração dos ecossistemas.</li>
    </ul>

<h3>4.1.4. Educação Ambiental e Conscientização</h3>
    <ul>
        <li>Campanhas para conscientizar agricultores sobre os impactos das queimadas e incentivar práticas mais sustentáveis, além de programas de educação ambiental em escolas e comunidades.</li>
    </ul>

<h3>4.1.5. Políticas de Incentivo à Conservação</h3>
    <ul>
        <li>Criar subsídios e incentivos fiscais para a preservação de áreas florestais, com punições mais rígidas para quem pratica queimadas ilegais.</li>
    </ul>

<h3>4.1.6. Mudanças Climáticas e Gestão de Recursos Hídricos</h3>
    <ul>
        <li>Desenvolver políticas de mitigação climática e manejo integrado do fogo para evitar grandes incêndios, promovendo o uso eficiente de recursos hídricos.</li>
    </ul>

<h3>4.1.7. Incentivo à Pesquisa e Inovação</h3>
    <ul>
        <li>Investir em tecnologias de combate ao fogo e práticas agrícolas mais resilientes, com foco em irrigação eficiente e monitoramento de condições climáticas.</li>
    </ul>

<h3>4.1.8. Colaboração Internacional</h3>
    <ul>
        <li>Estabelecer parcerias globais e cumprir acordos internacionais, como o Acordo de Paris, para proteger biomas e reduzir emissões de gases de efeito estufa.</li>
    </ul>
